```python
#SK97
# Defining a function that moves an object (or its contents) to a given device (like CPU or GPU)
def moveTo(obj, device):
    """
    obj: the Python object to move to a device, or to move its contents to a device
    device: the compute device to move objects to (e.g., 'cuda' for GPU)
    """
    if isinstance(obj, list):
        # If obj is a list, move each element in the list
        return [moveTo(x, device) for x in obj]
    elif isinstance(obj, tuple):
        # If obj is a tuple, convert to list, move elements, then convert back to tuple
        return tuple(moveTo(list(obj), device))
    elif isinstance(obj, set):
        # If obj is a set, convert to list, move elements, then convert back to set
        return set(moveTo(list(obj), device))
    elif isinstance(obj, dict):
        # If obj is a dictionary, move both keys and values
        to_ret = dict()
        for key, value in obj.items():
            to_ret[moveTo(key, device)] = moveTo(value, device)
        return to_ret
    elif hasattr(obj, "to"):
        # If obj has a .to() method (like a PyTorch tensor), move it to the device
        return obj.to(device)
    else:
        # If obj is something else, return it as is
        return obj

# Creating a list of two simple PyTorch tensors on CPU
some_tensors = [torch.tensor(1), torch.tensor(2)]

# Printing the original list of tensors (on CPU)
print(some_tensors)

# Moving the list of tensors to the given device (e.g., GPU) and printing the result
print(moveTo(some_tensors, device))
```

```python
#SK97
import matplotlib.pyplot as plt

# This code performs gradient descent using PyTorch to minimize the function f(x) = (x - 2)^2.
# It starts from x = -3.5 and iteratively updates x by moving against the gradient.
# The loop stops when the change between updates is smaller than epsilon.
# The expected result is that x will approach 2, where the function reaches its minimum.

# Creating a tensor with value -3.5 and enabling gradient tracking
x = torch.tensor([-3.5], requires_grad=True)

# Setting initial current value as a clone of x
x_cur = x.clone()

# Setting initial "previous" value far from x_cur to ensure the loop runs at least once
x_prev = x_cur * 100

# Setting a small epsilon for the stopping condition (tolerance)
epsilon = 1e-5

# Setting the learning rate (step size)
eta = 0.1

# Step counter and lists to store progress
step = 0
x_values = []
f_values = []

# Running gradient descent until the change between steps is smaller than epsilon
while torch.linalg.norm(x_cur - x_prev) > epsilon:
    # Store the current value as previous for the next check
    x_prev = x_cur.clone()  # Clone to avoid reference issues

    # Compute function value
    value = f(x)

    # Store progress
    x_values.append(x_cur.item())
    f_values.append(value.item())

    # Compute gradient
    value.backward()

    # Update x using gradient descent (x = x - eta * gradient)
    x.data -= eta * x.grad

    # Zero out the gradient for the next iteration (important in PyTorch)
    x.grad.zero_()

    # Update current value
    x_cur = x.data

    # Print progress
    step += 1
    print(f"Step {step}: x = {x_cur.item():.6f}, f(x) = {value.item():.6f}")

# Print the final optimized value (should approach x = 2, the minimum of f(x))
print(f"\nFinal result after {step} steps: x = {x_cur.item():.6f}")

# Plotting the progress
plt.figure(figsize=(10, 5))

# Plot x values over steps
plt.subplot(1, 2, 1)
plt.plot(range(1, step + 1), x_values, marker='o')
plt.title("x Value Over Steps")
plt.xlabel("Step")
plt.ylabel("x")

# Plot function values over steps
plt.subplot(1, 2, 2)
plt.plot(range(1, step + 1), f_values, marker='o', color='orange')
plt.title("f(x) Value Over Steps")
plt.xlabel("Step")
plt.ylabel("f(x)")

plt.tight_layout()
plt.show()
```

```python
#SK97
# Creating a PyTorch nn.Parameter, which is a special tensor used in neural networks
# It behaves like a tensor with requires_grad=True and is registered inside models
# This makes it easy to optimize using built-in optimizers like SGD
# This is useful if you want to optimize this parameter using
# PyTorch optimizers (like torch.optim.SGD) rather than manual gradient updates.
x_param = torch.nn.Parameter(torch.tensor([-3.5]), requires_grad=True)
```

```python
#SK97
# Creating an SGD (stochastic gradient descent) optimizer to update x_param
# We pass the parameter as a list [x_param] and set the learning rate (eta)
# With this, you no longer need to manually write x.data -= eta * x.grad.
optimizer = torch.optim.SGD([x_param], lr=eta)
```

```python
#SK97
# Running optimization for 60 epochs using the SGD optimizer
for epoch in range(60):
    # Clearing old gradients (this replaces x.grad.zero_())
    optimizer.zero_grad()

    # Calculating the current loss (f(x_param))
    loss_incurred = f(x_param)

    # Computing the gradient of the loss with respect to x_param
    loss_incurred.backward()

    # Updating x_param using the optimizer (this replaces manual update like x.data -= eta * x.grad)
    optimizer.step()

# Printing the final optimized value of x_param (should approach 2, the minimum of f(x))
print(x_param.data)
```

```python
#SK97
# Importing PyTorch's Dataset class (used to create custom datasets)
from torch.utils.data import Dataset

# Importing function to load datasets from OpenML
from sklearn.datasets import fetch_openml

# Loading the MNIST handwritten digits dataset (dataset ID 554) from OpenML
# X will contain the image data (as flattened 784-dimensional vectors)
# y will contain the labels (digits 0–9)
X, y = fetch_openml('mnist_784', version=1, return_X_y=True)

# Printing the shape of X (should be [70000, 784], meaning 70,000 images, each with 784 pixels)
print(X.shape)
```

```python
#SK97
# Defining a custom PyTorch dataset by subclassing Dataset
class SimpleDataset(Dataset):

    def __init__(self, X, y):
        # Calling the parent class constructor
        super(SimpleDataset, self).__init__()
        # Storing the input data (features) and labels (targets)
        self.X = X
        self.y = y

    def __getitem__(self, index):
        # Fetching one sample at the given index

        # Getting the feature values at this index, converting to float32 tensor
        inputs = torch.tensor(self.X.iloc[index, :].values, dtype=torch.float32)

        # Getting the label at this index, converting to int64 tensor
        targets = torch.tensor(int(self.y[index]), dtype=torch.int64)

        return inputs, targets

    def __len__(self):
        # Returning the total number of samples
        return self.X.shape[0]

# Creating a PyTorch dataset from the loaded MNIST data
dataset = SimpleDataset(X, y)
```
